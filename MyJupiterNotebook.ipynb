*creating a christmas tree*

```python
row = 7
def triangleShape(n):
    for i in range(n): #number of iterations
        for j in range(n-i): 
            print(' ', end=' ')  # what to do in every iteration: 1. print empty space
        for k in range(2*i+1):
            print('*',end=' ')   # 2. print * 2x plus 1 space ahead the iteration number
        print()
        
triangleShape(row)
triangleShape(row)
```
